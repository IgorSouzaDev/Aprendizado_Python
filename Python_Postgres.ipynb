```python
import pandas as pd
import sqlalchemy
```

Conectando ao Banco de dados
                                    #NomeDoBanco+driver//usuario:senha@host/banco


```python
engine=sqlalchemy.create_engine('postgresql+psycopg2://postgres:postgres@localhost/Integration')
```

Select da tabela

```python
fato_pedidos =pd.read_sql(sql= "SELECT * FROM fato_pedidos", con=engine)
dim_lojas =pd.read_sql(sql= "SELECT * FROM dim_lojas", con=engine)
dim_produtos =pd.read_sql(sql= "SELECT * FROM dim_produtos", con=engine) 
```

Merge das tabelas metodo LEFT JOIN

IN SQL = 
SELECT * FROM dim_produtos dim
LEFT JOIN fato_pedidos fat ON fat.produto =  dim.id

```python
fato_dw =pd.merge(left=fato_pedidos,right=dim_produtos,how='left', left_on='produto',right_on='id')
fato_dw
```

Merge das tabelas metodo LEFT JOIN
Como foi armazenado o Join anterior no python so é necessario fazer o novo join ao invez de colocar todos o join igual ao SQL

IN SQL = 

SELECT * FROM fato_pedidos fat
LEFT JOIN dim_produtos dim ON  dim.id = fat.produto
LEFT JOIN dim_lojas lj ON fat.loja = lj.id

```python
fato_dw =pd.merge(left=fato_dw,right=dim_lojas,how='left', left_on='loja',right_on='id')
fato_dw
```

```python
fato_dw.drop(columns=['id_x','id_y','produto_x','id'],inplace = True)

```

```python
fato_dw=fato_dw[['produto_y','valor','data','estado','logradouro','cidade']]
fato_dw
```

```python
fato_dw.rename(columns={"produto_y": "Produto"})
```
